```python
import pandas as pd
```

```python
maturities=['WGS3MO.csv','WGS6MO.csv','WGS1YR.csv','WGS2YR.csv','WGS3YR.csv','WGS5YR.csv','WGS7YR.csv','WGS10YR.csv','WGS20YR.csv','WGS30YR.csv']
df=pd.read_csv('WGS1MO.csv',index_col=0)
for i in maturities:
    df1=pd.read_csv(i,index_col=0)
    df=pd.concat([df, df1], axis=1).reindex(df1.index)
df.columns=['1m','3m','6m','1yr','2yr','3yr','5yr','7yr','10yr','20yr','30yr']
df
```

```python
df.to_csv('CMT.csv')
```
